### Data Manipulation

1. Handling Null Values
2. Slicing / Selecting
3. Filtering

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

```python
# fillna - filling the missing values
#Fill with mean,median or mode for numerical data
#Fill mode in categorical data

data['Item_Weight'].fillna(data['Item_Weight'].mean(),inplace=True)
```

```python
data['Outlet_Size'].fillna(data["Outlet_Size"].mode()[0],inplace=True)
```

```python
data['Outlet_Location_Type'].fillna(data['Outlet_Location_Type'].mode()[0],inplace = True)
```

```python
data.isnull().sum()
```
